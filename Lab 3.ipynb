Part A : Classical Computer Vision (OpenCV)

```python
# Using OpenCV write a function which detects and tracks the center of a white line in an image.

import cv2
import numpy as np
import matplotlib.pyplot as plt

def detect_and_track_line(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY)
    
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if contours:
        largest_contour = max(contours, key=cv2.contourArea)
        
        # Fit a line to the contour points
        [vx, vy, x0, y0] = cv2.fitLine(largest_contour, cv2.DIST_L2, 0, 0.01, 0.01)
        vx, vy, x0, y0 = vx[0], vy[0], x0[0], y0[0]  # convert to scalars
        
        center = (int(x0), int(y0))
        angle = np.arctan2(vy, vx) * 180 / np.pi
        
        return center, angle, (vx, vy)
    
    return None, None, None

# Process multiple images
for n in range(1, 4):
    image = cv2.imread(f'White_Line_{n}.png')
    
    plt.figure(figsize=(8,6))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f'Image {n} Before Line Detection')
    plt.axis('off')
    plt.show()

    center, angle, direction = detect_and_track_line(image)
    print(f"Image {n}: Center = {center}, Angle = {angle:.2f} degrees")

    if center:
        cv2.circle(image, center, 5, (0, 255, 0), -1)

        # Draw long line in both directions
        vx, vy = direction
        height, width = image.shape[:2]
        length = max(width, height)  # make sure it crosses the whole image

        # Start and end points along the line
        x1 = int(center[0] - vx * length)
        y1 = int(center[1] - vy * length)
        x2 = int(center[0] + vx * length)
        y2 = int(center[1] + vy * length)

        cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 2)

    plt.figure(figsize=(8,6))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f'Line Detection in Image {n}')
    plt.axis('off')
    plt.show()


    
```

```python
import cv2
import matplotlib.pyplot as plt
import time
import cv2

# Create a video capture object, usually 0 for the default camera
cap = cv2.VideoCapture(0)

# Check if the camera opened successfully
if not cap.isOpened():
    print("Could not open video device")
else:
    print("Webcam opened successfully. Press 'q' to exit.")

n = 0
ret, frame = cap.read()
frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(frame_rgb)
plt.axis('off')

while n<(30*5):  # Run for approximately 5 seconds at 30 FPS
    # Read a frame from the webcam
    ret, frame = cap.read()

    # If the frame was not successfully captured, break the loop
    if not ret:
        break

    # Display the captured frame in a window
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Show frame
    #plt.imshow(frame_rgb)
    #plt.axis('off')
    plt.show()
    plt.pause(0.001)  # Pause to allow the plot to update
    plt.clf()  # Clear the figure for the next frame

    # Break the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    n += 1

# Release the video capture object and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

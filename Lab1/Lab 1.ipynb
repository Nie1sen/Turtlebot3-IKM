Why is numpy faster than using for loops in Python for operations such as matrix
multiplications?

Python loops do math one step at a time with lots of multiplication, while NumPy does the whole thing in pre-made, low level, optimized code using vectors instead of individual elements for matrix multiplication.

What is the data type used in numpy which makes such operations feasible? Also name a few
differences between this data type and its similar counterpart in Python.



A NumPy array stores elements as vectors in a continuous block of memory as one, raw data type. Because they use vectors, they significanlty simplify the matrix computations as opposed to elementwise multiplication and addition in Python using for loops. Python lists on the other hand, store references to separate Python objects scattered throughout memory so its slower to access and compute.

Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a line plot of the sine function over the interval [0, 2π]
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.plot(x, y)

# Add labels to the axes in a Matplotlib plot.

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sine Function')

# Plot the 3d graph of the function given below using Matplotlib
# Z = sin(sqrt(X^2 + Y^2))
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()
```

SciPy

```python
from scipy.linalg import solve

# Solve the linear system of equations given below using SciPy.
# 3𝑥 + 𝑦 = 9
# 𝑥 + 2𝑦 = 8

A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])
solution = solve(A, b)
print(f'The solution is x = {solution[0]:.2f} and y = {solution[1]:.2f}')

#Find the minimum of the function given below using SciPy's optimization module. 𝑦 = 𝑥2 + 2𝑥
from scipy.optimize import minimize
def objective_function(x):
    return x**2 + 2*x
result = minimize(objective_function, x0=0)  # Starting point for optimization
print(f'The minimum value of the function is {result.fun:.2f} at x = {result.x[0]:.2f}')


# Perform the Fourier transformation of the function given below using SciPy. Plot the frequency response using matplotlib.
# f(x) = sin(100πx) + 1/2 sin(160πx)
from scipy.fft import fft, fftfreq
# Define the function
def f(x):
    return np.sin(100 * np.pi * x) + 0.5 * np.sin(160 * np.pi * x) 
# Sample the function
x = np.linspace(0, 1, 1000)
y = f(x)
# Perform the Fourier transformation
yf = fft(y)
xf = fftfreq(len(x), x[1] - x[0])
# Plot the frequency response
plt.plot(xf, np.abs(yf))
plt.title('Frequency Response of f(x)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.xlim(0, 200)  # Limit x-axis to show relevant frequencies
plt.grid()
plt.show()
```

```python
import cv2
# Read the image
image = cv2.imread('Homelander.jpg')
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Perform edge detection on the image using OpenCV
edges = cv2.Canny(gray_image, 100, 200)
# Use a Haar cascade classifier to implement face detection using OpenCV on an image which contains faces.
# Load the Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
# Detect face in the image
face = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5)
# Draw rectangles around the detected face
for (x, y, w, h) in face:
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)
# Display the image with detected face
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Face Detection')
plt.axis('off')
```

Explain how Haar cascades work to detect faces.

Haar cascades detect faces by looking for simple patterns of light and dark in an image. The system uses small black and white rectangles called Haar features to find edges and lines that are common on faces, like the eyes, nose, and mouth. It uses an integral image, which lets it quickly sum pixel values in any rectangular area. The system is trained on many face and non-face images so it learns which features are most useful for detecting faces. When detecting, it moves a small window over the image and uses a cascade of classifiers that first checks simple features to reject non-face areas quickly and then checks more detailed features on the remaining areas. This allows Haar cascades to find faces efficiently and in real time.
